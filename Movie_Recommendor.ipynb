```python
import json
def genres_display(json_str):
    L = []
# Loading the JSON string into a list of dictionaries
    genres_list = json.loads(json_str)
    for genre in genres_list:
        L.append(genre['name'])
    return L

# Apply the function to the 'genres' column
movies['genres'] = movies['genres'].apply(genres_display)
```

```python
def display_keywords(json_str):
    L = []
    for i in json.loads(json_str):
        L.append(i['name'])
    return L
movies['keywords'] = movies['keywords'].apply(display_keywords)
```

```python
# from this above dic director name , producer  , writer names are important in selecting the movie preferences
def crew_detail(json_str):
    L = []
    count = 0
    for crew in json.loads(json_str):
        if crew['job'] == 'Director' :
            L.append(crew['name'])
            count+=1
        if count == 1:
            break
    return L
```

```python
def crew_job(json_str):
    l = []
    for crew in json.loads(json_str):
        l.append(crew['job'])
    return l
```

# Modal preparation

```python
# using vectorization of text of the description column, purpose is to match the number of words of one movie relative to each movie one by one
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 5000 , stop_words = 'english' )
```

```python
# now the problem is countvectorizer object fail to understand , that games , game has same menaing, or in between race and racing so to deal with it i am steaming the word
import nltk

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def steam(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
movies['Description'] = movies['Description'].apply(steam)
```

```python
vectors = cv.fit_transform(movies['Description']).toarray()
```

```python
vectors[0]
```

```python
np.set_printoptions(threshold=np.inf)
cv.get_feature_names_out()
```

```python
# calculating the angle between the vector of each movie realtive to every another movie
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
# My aim is to select top 5 closest value , which wiil be nearer to the given movie value
def recommend(movie):
    movie_index =  movies[movies.title == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(similarity[movie_index])) , reverse = True , key = lambda x: x[1])[1:6]

    for i in movies_list:
        print(movies.iloc[i[0]].title)
    
```

```python
import requests

def fetch_poster(movie_id):
    try:
        api_key = '7650cbf317132270032f652bab2cfa7b'  # Replace with your actual TMDB API key
        url = f'https://api.themoviedb.org/3/movie/{movie_id}?api_key={api_key}'
        
        response = requests.get(url)
        if response.status_code == 200:
            data = response.json()
            if 'poster_path' in data:
                image = data['poster_path']
                return 'https://image.tmdb.org/t/p/w500/' + image
            else:
                print(f"No poster path found for movie ID: {movie_id}")
                return None
        else:
            print(f"Failed to fetch data for movie ID {movie_id}. Status code: {response.status_code}")
            return None
    except Exception as e:
        print(f"Error fetching poster for movie ID {movie_id}: {e}")
        return None
```
